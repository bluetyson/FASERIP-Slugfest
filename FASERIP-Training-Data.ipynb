```python
def getrank(rankstr):
    #print(rankstr.strip())
    rank = rankstr.strip().split(' ')[-1]
    if 'Sh' in rank:
        rank = rankstr.strip().split(' ')[-1][0:4]
        rank = rank.replace('-','')
    else:
        rank = rankstr.strip().split(' ')[-1][0:2]
    #print(rank)
    return rank

def getHK(rankstr):
    #print(rankstr.strip())
    rank = rankstr.strip().split(' ')
    
    if len(rank) < 4:
        return [0, 0]
    else:

        Health = rank[1]
        Karma = rank[3]
    #print(rank)
    return [Health, Karma]

def getTalents(rankstr):
    Talents = rankstr.strip().replace(rankstr,'')
    return Talents
```
